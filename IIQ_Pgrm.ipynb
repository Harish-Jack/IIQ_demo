# Sum Of Two Numbers

```python
a = int(input("enter first number: "))
b = int(input("enter second number: "))
if a > 0:

    if b>0:
    
      sum = a + b
      print("sum:", sum)  
    
    else:
    
        print("The First Number is 0 so unable to perform")
    
    

else:
    
        print("The Second Number is 0 so unable to perform")
```
